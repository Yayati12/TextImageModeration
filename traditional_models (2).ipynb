```python
import re
import emoji
from indicnlp.tokenize import indic_tokenize

# Load stopwords
def load_stopwords(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        stop_words = set(f.read().splitlines())
    return stop_words

# Emoji to text
def convert_emojis_to_text(text):
    return emoji.demojize(text, language='en')

# Clean text
def clean_text(text):
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"[^‡§Ä-‡•øa-zA-Z:_ ]+", "", text)
    text = convert_emojis_to_text(text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


# Remove stopwords
def remove_stopwords(text, stop_words):
    tokens = indic_tokenize.trivial_tokenize(text)
    return ' '.join([word for word in tokens if word not in stop_words])

# Full pipeline
def preprocess_pipeline(text, stop_words, remove_stops=True):
    text = clean_text(text)
    if remove_stops:
        text = remove_stopwords(text, stop_words)
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

# Fit on train set and transform all
X_train_tfidf = tfidf_vectorizer.fit_transform(train['text'])
X_test_tfidf = tfidf_vectorizer.transform(test['text'])
X_valid_tfidf = tfidf_vectorizer.transform(valid['text'])

# Get labels
y_train = train['label']
y_test = test['label']
y_valid = valid['label']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Initialize and train the model
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_tfidf, y_train)
```

```python
def evaluate_model(name, model, X_train, y_train, X_val, y_val, X_test, y_test):
    print(f"\n {name} - TF-IDF")

    # Train accuracy
    y_pred_train = model.predict(X_train)
    print("Train Accuracy:", accuracy_score(y_train, y_pred_train))
    print("Train Report:\n", classification_report(y_train, y_pred_train))

    # Validation accuracy
    y_pred_val = model.predict(X_val)
    print("Validation Accuracy:", accuracy_score(y_val, y_pred_val))
    print("Validation Report:\n", classification_report(y_val, y_pred_val))

    # Test accuracy
    y_pred_test = model.predict(X_test)
    print("Test Accuracy:", accuracy_score(y_test, y_pred_test))
    print("Test Report:\n", classification_report(y_test, y_pred_test))

```

```python
# Evaluate Logistic Regression
evaluate_model("Logistic Regression", lr_model, X_train_tfidf, y_train, X_valid_tfidf, y_valid, X_test_tfidf, y_test)
```

```python
from sklearn.svm import LinearSVC

# Initialize and train the model
svm_model = LinearSVC()
svm_model.fit(X_train_tfidf, y_train)

```

```python
# Evaluate SVM
evaluate_model("SVM", svm_model, X_train_tfidf, y_train, X_valid_tfidf, y_valid, X_test_tfidf, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_tfidf, y_train)
```

```python
# Evaluate Random Forest
evaluate_model("Random Forest", rf_model, X_train_tfidf, y_train, X_valid_tfidf, y_valid, X_test_tfidf, y_test)
```

```python
from gensim.models import Word2Vec
import numpy as np
```

```python
# Tokenize sentences into words
train_sentences = [sentence.split() for sentence in train['text']]

# Train Word2Vec
w2v_model = Word2Vec(sentences=train_sentences, vector_size=100, window=5, min_count=2, workers=4, sg=1, seed=42)

# Save word vector size for later use
w2v_dim = w2v_model.vector_size
```

```python
# Function to average word vectors for a sentence
def get_sentence_vector(sentence, model, dim):
    words = sentence.split()
    word_vecs = [model.wv[word] for word in words if word in model.wv]
    if word_vecs:
        return np.mean(word_vecs, axis=0)
    else:
        return np.zeros(dim)

# Apply to all datasets
X_train_w2v = np.array([get_sentence_vector(text, w2v_model, w2v_dim) for text in train['text']])
X_valid_w2v = np.array([get_sentence_vector(text, w2v_model, w2v_dim) for text in valid['text']])
X_test_w2v  = np.array([get_sentence_vector(text, w2v_model, w2v_dim) for text in test['text']])
```

```python
def evaluate_model1(name, model, X_train, y_train, X_val, y_val, X_test, y_test):
    print(f"\n {name} - Word2Vec")

    # Train accuracy
    y_pred_train = model.predict(X_train)
    print("Train Accuracy:", accuracy_score(y_train, y_pred_train))
    print("Train Report:\n", classification_report(y_train, y_pred_train))

    # Validation accuracy
    y_pred_val = model.predict(X_val)
    print("Validation Accuracy:", accuracy_score(y_val, y_pred_val))
    print("Validation Report:\n", classification_report(y_val, y_pred_val))

    # Test accuracy
    y_pred_test = model.predict(X_test)
    print("Test Accuracy:", accuracy_score(y_test, y_pred_test))
    print("Test Report:\n", classification_report(y_test, y_pred_test))

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC  # SVM classifier
```

```python
# Initialize and train
lr_w2v_model = LogisticRegression(max_iter=1000)
lr_w2v_model.fit(X_train_w2v, y_train)

# Evaluate on all sets
evaluate_model1("Logistic Regression", lr_w2v_model,
               X_train_w2v, y_train, X_valid_w2v, y_valid, X_test_w2v, y_test)
```

```python
# Train SVM on Word2Vec vectors
svm_w2v_model = LinearSVC()
svm_w2v_model.fit(X_train_w2v, y_train)

# Evaluate on all sets
evaluate_model1("SVM", svm_w2v_model,
               X_train_w2v, y_train, X_valid_w2v, y_valid, X_test_w2v, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train
rf_w2v_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_w2v_model.fit(X_train_w2v, y_train)

# Evaluate on all sets
evaluate_model1("Random Forest", rf_w2v_model,
                X_train_w2v, y_train,
                X_valid_w2v, y_valid,
                X_test_w2v, y_test)
```
